```python
# 해당 문자열이 알파벳만으로 이루어져 있는지 확인하기 위한 .isalpha
# (단 단순히 영문 알파벳만을 뜻하는 것이 아니라 유니코드를 기준으로 각 나라의 단어를 구성하는 문자라는 의미에서의 알파벳)

a = 'python'
b = 'python3.6'
c = 'python!'
d = 'python is fun'
e = '파이썬'

a.isalpha(), b.isalpha(), c.isalpha(), d.isalpha(), e.isalpha()
```

```python
# 해당 문자열이 숫자로만 이루어져 있는지를 검토하는 .isdigit()
a = '1234'
b = '1234a'
c = '12 34'

a.isdigit(), b.isdigit(), c.isdigit()
```

```python
# 해당 문자열이 특정 단어로 시작하는지 여부를 체크하는 .startswith()
a = 'Life is to short'
a.startswith('Life'), a.startswith('life'), a.startswith('L'), a.startswith('short')
```

```python
# 해당 문자열이 특정 단어로 끝나는지 여부를 체크하는 .endswith()
a = 'Life is to short'
a.endswith('Life'), a.endswith('life'), a.endswith('ort'), a.endswith('short')
```
