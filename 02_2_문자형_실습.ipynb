```python
'''
"Error is %d%." % 98 라고 입력할 경우 incomplete format 오류가 발생한다.
"Error is %d%%." % 98 와 같이 포매팅 연산자를 사용할 경우 %를 출력하기 위해서는 %%로 나타내야 한다.
'''

"Error is %d%%." %98
```
